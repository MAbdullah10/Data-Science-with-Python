### Multiple Linear Regression

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
import pandas as pd
df = pd.read_csv('Salary Data.csv')
df.head()
```

```python
x = df[['Age','Years of Experience']]
y = df['Salary']
```

```python
df.isnull().sum()
```

```python
x = x.dropna()
y = y.dropna()
```

```python
x.isnull().sum()
```

```python
# create and fit the model
model= LinearRegression().fit(x,y)
model
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[32,5.0]])
```

```python
model.score(x,y)
```

We have directly trained the model first now we will split the data into training and testing and then comapre their scores

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)
```

```python
model = LinearRegression()
model = model.fit(x_train,y_train)
model
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
model.predict([[28,3.0]])
```

#### R Squared Value

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import  r2_score

r2 = r2_score(y_test, y_pred)
print(f"R^2 Score: {r2}")
```

#### Plotting

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(x_train.iloc[:, 0], y_train, color='blue', label='Training Data')
plt.plot(x_train.iloc[:, 0], model.predict(x_train), color='green', label='Regression Line')

plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(x_test.iloc[:, 0], y_test, color='blue', label='Training Data')
plt.plot(x_test.iloc[:, 0], model.predict(x_test), color='green', label='Regression Line')

plt.legend()
plt.show()
```
